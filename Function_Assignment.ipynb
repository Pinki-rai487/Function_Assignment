6. Explain the concept of generators in Python and how they are defined ?

A generator function is a special type of function that returns an iterator object. Instead of using return to send back a single value, generator functions use yield to produce a series of results over time. This allows the function to generate values and pause its execution after each yield, maintaining its state between iterations.

```python
#6  example of generator:

def Num(max):
  count = 1
  while count <= max:
    yield count
    count += 1

X = Num(5)
for i in X:
   print(i)
```

7. What are the advantages of using generators over regular functions ?

Ans>

1.  Memory Efficiency: One of the biggest advantages of generators is that they are memory efficient. Since a generator yields its values one by one, it does not need to store all of the values in memory at once.
2.  Readability: Generators can make your code easier to read and understand. By breaking up the iteration process into smaller chunks.
3. Speed: Generators can also be faster than regular loops, especially when working with large data sets.

```python
#4. Write a Python function that checks if a given number is prime or not from 1 to 200 ?
def is_prime(num):
  if num <=1:
    return False
  for i in range(2,int(num**0.5)+1):
    if num % i == 0:
      return False
  return True

def prime_generator():
  for num in range(1,201):
    if is_prime(num):
      yield num
    num +=1

gen = prime_generator()
for _ in range(200):
   print("It is prime number",next(gen))
   break
print(next(gen))
print(next(gen))
```

```python
#6. Write a generator function in Python that yields the powers of 2 up to a given exponent ?

def calculate(exponent):
  for i in range(1,exponent + 1):
    yield 2**i

exp_Num = 5
powers = calculate(exp_Num)

print(f"Powers of 2 up to exponent {exp_Num} using next():")
print(next(powers))
print(next(powers))
print(next(powers))
print(next(powers))
print(next(powers))
```

```python
#7.  Implement a generator function that reads a file line by line and yields each line as a string ?

import os

def read_file_lines(filename):
  if not os.path.exists(filename):
    print(f"Error: File '{filename}' not found.")
    return

  with open(filename, 'r') as f:
    for line in f:
      yield line.strip()

file_content = """This is the first line.
This is the second line.
And this is the third line."""

file_name = "my_document.txt"

with open(file_name, "w") as f:
  f.write(file_content)

print(f"Reading lines from '{file_name}':")

line_generator = read_file_lines(file_name)
for line in line_generator:
  print(line)
os.remove(file_name)
```

```python
#8.  Use a lambda function in Python to sort a list of tuples based on the second element of each tuple ?

list_of_tuples = [(1, 5, 10), (3, 2), (2, 8, 9, 12), (4, 1, 6)]
sorted_list = sorted(list_of_tuples, key=lambda item: item[1])

print(f"Original list: {list_of_tuples}")
print(f"Sorted list: {sorted_list}")
```
